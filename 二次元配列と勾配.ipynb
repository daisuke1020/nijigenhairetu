```python
import numpy as np

x_ndarray =np.arange(-50,50.1,0.1)  #(strat,stop,step)
y_ndarray =[]                       #空のリストを作成

for i in range(0,1001):
    y_ndarray.append(x_ndarray[i]/2 +1)

y_ndarray = np.array(y_ndarray)    #Yリストをnpリストにする


z_ndarray = np.vstack([x_ndarray,y_ndarray])    #XリストとYリスとを結合
z_ndarray = z_ndarray.T                         #Zリストを転置

a_ndarray =[]
for j in range(1,1001):
    a_ndarray.append((z_ndarray[j,1:] - z_ndarray[j-1,:-1])/0.1)  #z[j,0]-z[j-1,1]のループ

a_ndarray = np.array(a_ndarray)
a_ndarray = a_ndarray.T
a_ndarray = np.round(a_ndarray,2)
print(a_ndarray)    #zリストの全ての勾配が代入されたリスト
```
